```python

ps = PorterStemmer()
af = Afinn()
stop_words = set(stopwords.words('english'))

stop_words_no_afinn = set([w for w in stop_words if af.score(w)==0])

print(len(stop_words))
print(len(stop_words_no_afinn))
#count = lambda l1, l2: len(list(filter(lambda c: c in l2, l1)))
```

```python
auth_files_ls = []
for author in os.listdir(data_dir):
    if author == 'AaronPressman':
        for files in os.listdir(os.path.join(data_dir, author)):
            _filename = os.path.join(data_dir, author, files)
            with open(_filename, 'r') as f:
                txt = f.read()
            total_sents = len(nltk.sent_tokenize(txt))
            total_tokens_raw = len(word_tokenize(txt))
            total_tokens_lower = len(word_tokenize(txt.lower()))
            unique_tokens_lower = len(set(word_tokenize(txt.lower())))
            unique_tokens_lower_stem = len(set([ps.stem(w) for w in word_tokenize(txt.lower())]))
            unique_tokens_lower_stem_no_stopwords = len(set([ps.stem(w) for w in word_tokenize(txt.lower())  if not w in stop_words_no_afinn]))
            total_stopwords = len([stop_words_no_afinn for w in word_tokenize(txt.lower())])
            avg_word_sent = (total_tokens_raw/total_sents)
            avg_word_len = np.mean([len(w) for w in word_tokenize(txt.lower())  if w not in stop_words_no_afinn])
            sentiment_score = af.score(txt)
            #total_punct = count(txt, string.punctuation)
            total_punct = len(re.sub('[a-z0-9\s]', '', txt, flags=re.I|re.M))
            total_nouns = len([token for token, pos in pos_tag(word_tokenize(txt)) if pos.startswith('N')])
            total_verb = len([token for token, pos in pos_tag(word_tokenize(txt)) if pos.startswith('V')])
            total_adjective = len([token for token, pos in pos_tag(word_tokenize(txt)) if pos.startswith('J')])
            auth_files_ls.append({
                'total_sents': total_sents,
                'total_tokens_raw': total_tokens_raw,
                'total_tokens_lower': total_tokens_lower,
                'unique_tokens_lower': unique_tokens_lower,
                'unique_tokens_lower_stem': unique_tokens_lower_stem,
                'unique_tokens_lower_stem_no_stopwords' : unique_tokens_lower_stem_no_stopwords ,
                'total_stop_words' :  total_stopwords,
                 'avg_word_sent' : avg_word_sent ,
                 'avg_word_len' : avg_word_len ,
                 'sentiment_score': sentiment_score ,
                'total_punct': total_punct,
                 'total_nouns' :total_nouns ,
                 'total_verb':total_verb ,
                 'total_adjective':total_adjective,
                'Author' : author
            })
auth_df = pd.DataFrame(auth_files_ls)
```

```python
from sklearn.model_selection import train_test_split , KFold , cross_val_score
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score , recall_score , classification_report
from sklearn.ensemble import RandomForestClassifier
```

```python
X = auth_df.drop('Author',axis=1)
y = auth_df["Author"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=7)
```

```python
rf = RandomForestClassifier(criterion='gini')
```

```python
rf.fit(X_train,y_train)
y_pred_rf = rf.predict(X_test)
```

```python
print('accuracy_score',accuracy_score(y_test,y_pred_rf))
print('precision_score',precision_score(y_test,y_pred_rf,average=None))
print('recall_score',recall_score(y_test,y_pred_rf,average=None))
#print(confusion_matrix(y_test,y_pred_rf))
print(classification_report(y_test,y_pred_rf))
```
